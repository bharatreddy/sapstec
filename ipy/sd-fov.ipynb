```python
from davitpy.pydarn.radar import *
from davitpy.pydarn.plotting import *
from davitpy.utils import *
import datetime as dt
%pylab inline
```

```python
# Set map
fig = figure(figsize=(10,10))
m = plotUtils.mapObj(boundinglat=30., coords='mag')
radars = network()
codes = radars.getAllCodes()
markcodes = ['wal','fhe','fhw','cve','cvw','ade','adw','bks']
# Plotting some radars
overlayRadar(m, fontSize=12, codes=codes)
# Plot radar fov
overlayFov(m, codes=markcodes, maxGate=70, fovColor=(.8,.7,.8), fovAlpha=.5)
overlayFov(m, codes=codes[:-1], maxGate=70)#, fovColor=(.8,.9,.9))

fig.tight_layout(pad=2)
rcParams.update({'font.size': 12})
fig.savefig("../figs/fovs-highlighted.pdf",bbox_inches='tight')
```
